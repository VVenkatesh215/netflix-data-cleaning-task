### Identify and handle missing values using .isnull() in Python


```python
print("Missing values per column:\n",df.isnull().sum())
```

```python
df['director'].fillna('Unknown')
df['cast'].fillna('Unknown')
df['country'].fillna('Unknown')
df['date_added'].fillna('Not Available')
# Fill missing 'rating' with most frequent value (mode)
df['rating'].fillna(df['rating'].mode()[0])
```

```python
df.isnull().sum()
```

### Remove duplicate rows using .drop_duplicates()

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

### Standardize text values like gender, country names, etc.


```python
df.info()
```

```python
df.dtypes
```

```python
df['rating'].dtype
```

```python
df['type']=df['type'].str.strip().str.title()
df['country']=df['country'].str.strip().str.title()
df['rating']=df['rating'].str.strip().str.upper()
```

### Convert date formats to a consistent type (e.g., dd-mm-yyyy).


```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')

df['date_added']=df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
df.head(10)
```

### Rename column headers to be clean and uniform (e.g., lowercase, no spaces).


```python
df.columns=df.columns.str.strip().str.lower().str.replace(' ','_')
```
